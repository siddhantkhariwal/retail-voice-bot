```python
from docx import Document


def extract_docx_content(file_path):
    doc = Document(file_path)
    
    toc = []
    headings = {}
    
    for paragraph in doc.paragraphs:
        # Check for table of contents by looking for specific styles
        if paragraph.style.name == 'TOC Heading':
            toc.append(paragraph.text)
        elif paragraph.style.name.startswith('Heading'):
            heading_level = paragraph.style.name.split(' ')[-1]  # Get the heading level (e.g., '1', '2', ...)
            headings[heading_level] = headings.get(heading_level, []) + [paragraph.text]
    
    return toc, headings

def print_content(toc, headings):
    print("Table of Contents:")
    for item in toc:
        print(f"- {item}")
    
    print("\nHeadings:")
    for level, titles in headings.items():
        print(f"Heading Level {level}:")
        for title in titles:
            print(f"  - {title}")

    
```

```python
from pdf2image import convert_from_path
import os

# Function to convert PDF to images
def pdf_to_images(pdf_path, output_folder, dpi=300):
    # Convert PDF pages to images
    images = convert_from_path(pdf_path, dpi)

    # Ensure output folder exists
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Save each page as an image file
    for i, image in enumerate(images):
        image_path = os.path.join(output_folder, f'page_{i + 1}.png')
        image.save(image_path, 'PNG')
        print(f'Saved: {image_path}')

# Example usage
pdf_path = 'India Employee Handbook - 2024.pdf'  # Path to the input PDF
output_folder = 'output_images'  # Folder to save the images

pdf_to_images(pdf_path, output_folder)
```

```python
import pdfplumber
from embeddings.create_embedding import embed_text
# embeddings = model.get_embeddings(texts)

# Function to extract text from PDF
def extract_text_from_pdf(pdf_path):
    # Open the PDF file
    with pdfplumber.open(pdf_path) as pdf:
        all_text = ""
        extract = []
        # Iterate over all the pages
        for i, page in enumerate(pdf.pages):
            text = page.extract_text()  # Extract text from each page
            embed = embed_text(text)
            extract.append([text,embed[0]])
            if text:
                all_text += text
            print(f"Extracted text from page {i + 1}")

        return extract

# Example usage
pdf_path = 'data/India Employee Handbook - 2024.pdf'  # Path to the input PDF
extracted_text = extract_text_from_pdf(pdf_path)

# # Save the extracted text to a file
# with open('extracted_text.txt', 'w') as text_file:
#     text_file.write(extracted_text)

print("Text extraction complete. Check extracted_text.txt for the output.")
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# Function to compute cosine similarity and return top N most similar vectors
def cosine_similarity_search(vectors, target_vector, top_n=3):
    # Reshape target_vector to 2D since cosine_similarity expects 2D arrays
    target_vector = np.array(target_vector).reshape(1, -1)
    
    # Compute cosine similarities between the target vector and all other vectors
    similarities = cosine_similarity(vectors, target_vector)
    
    # Flatten the result array
    similarities = similarities.flatten()

    # Get indices of top N most similar vectors (in descending order of similarity)
    top_n_indices = similarities.argsort()[::-1][:top_n]
    
    # Get corresponding similarity scores
    top_n_similarities = similarities[top_n_indices]
    
    return top_n_indices, top_n_similarities

# Example dataset of vectors (these can be word embeddings, document embeddings, etc.)
# vectors = np.array([
#     [0.1, 0.2, 0.3],
#     [0.9, 0.8, 0.7],
#     [0.4, 0.5, 0.6],
#     [0.2, 0.2, 0.2]
# ])
vectors = np.array(vector)

# Target vector to search for
target_vector = np.array(embed_text("hello"))

# Perform cosine similarity search for top 3 similar vectors
top_indices, top_similarities = cosine_similarity_search(vectors, target_vector, top_n=3)

# Print the top 3 most similar vectors and their cosine similarities
print("Top 3 most similar vectors:")
for i, (index, similarity) in enumerate(zip(top_indices, top_similarities)):
    # print(f"Rank {i + 1}: Vector at index {index} with cosine similarity {similarity:.4f}")
    # print(f"Vector: {vectors[index]}\n")
    print(index, similarity)

```

```python
import numpy as np
from sklearn.neighbors import NearestNeighbors

# Function to perform KNN similarity search using cosine similarity
def knn_similarity_search(vectors, target_vector, k=3):
    # Initialize the NearestNeighbors model with cosine similarity
    knn_model = NearestNeighbors(n_neighbors=k, metric='cosine')
    
    # Fit the model on the dataset of vectors
    knn_model.fit(vectors)
    
    # Reshape target vector to 2D (as required by scikit-learn)
    target_vector = np.array(target_vector).reshape(1, -1)
    
    # Perform the KNN search
    distances, indices = knn_model.kneighbors(target_vector)
    
    # Return the indices of the nearest vectors and their distances
    return indices.flatten(), distances.flatten()

# Example dataset of vectors (these can be document embeddings, word embeddings, etc.)
vectors = np.array(vector)

# Target vector to search for
target_vector = np.array(embed_text("hello"))

# Perform KNN similarity search to find top 3 most similar vectors
top_indices, top_distances = knn_similarity_search(vectors, target_vector, k=3)

# Convert cosine distance to cosine similarity (cosine similarity = 1 - cosine distance)
top_similarities = 1 - top_distances

# Print the top 3 most similar vectors and their cosine similarities
print("Top 3 most similar vectors:")
for i, (index, similarity) in enumerate(zip(top_indices, top_similarities)):
    # print(f"Rank {i + 1}: Vector at index {index} with cosine similarity {similarity:.4f}")
    # print(f"Vector: {vectors[index]}\n")
    print(index, similarity)
```
